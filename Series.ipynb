```python
# A Series in Pandas is: 👉 A 1D labeled array(VERTICAL) NOT WORK WITH 2D ARRAY
# Means:
# values hote hain
# aur har value ka ek label/index hota hai
```

```python
import pandas as pd 
import numpy as np
```

```python
labels=['a','b','c']
my_list=[10,20,30]
arr=np.array([10,20,30])
dictionary={1:10,2:20,3:30}
```

```python
# DEFAULT LABELS/INDEX
pd.Series(my_list)
```

```python
# DYNAMIC LABEL/INDEXES
pd.Series(my_list,index=labels)
```

```python
pd.Series(arr)
```

```python
pd.Series(dictionary)
```
