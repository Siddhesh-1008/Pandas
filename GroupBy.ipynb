```python
import pandas as pd

data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}

df = pd.DataFrame(data)

print(df)
```

```python
# GROUP BY BASED ON CATEGORY AND THEN SUM THE SALES IN EACH CATEGORY
# Group By is performed on category 
df.groupby("Category")
```

```python
df.groupby("Category")['Sales'].sum()
```

```python
df2=df.groupby("Store")
# df2 GIVES DATAFRAME
df2['Sales'].sum()
```

```python
# SUM OF SALES PERFROMED ON BOTH CATEGORY and Store
df.groupby(['Category','Store'])['Sales'].sum()
```

```python
 # AGGREGAE FUNCITON mean max sum() meedian are used with group by
df['Sales'].sum()
df['Sales'].mean()
df['Sales'].median()
df['Sales'].mode()
df['Sales'].count()
```
