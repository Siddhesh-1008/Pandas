```python
import pandas as pd
import numpy as np

# Create DataFrame
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South'] * 5,
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice'] * 5
}

df = pd.DataFrame(data)

# Add Month Column
df['Month'] = df['Date'].dt.month_name()

# Add Quarter Column
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)

print(df)
```

```python
# values konse data pae karna hae index basically converts index data into rows columns basicaly converts product inside data into columns
# and aggfunc basically perform aggregate functino on dataframe
# (dataframe,values=data,index=column ko row mae convert karo,columns=column_name,aggfunc=all aggfunc)
# IF U WNAT DATA FROM MULTIPLE COLUMNS THEN MAKE USE OF values=["row1","row2"] AND ALWAYS VALUES MUST BE NUMERIC VALUE
pd.pivot_table(
    df,
    values='Sales',
    index='Region',
    columns='Product',
    aggfunc='sum'
)
```
