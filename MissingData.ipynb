```python
import numpy as np
import pandas as pd
```

```python
data1 = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df1=pd.DataFrame(data1)
```

```python
# TO KONOW HOW MANY TRUE AND HOW MANY FALSE VALUE 
df1.isna()
```

```python
# TO KONOW HOW MANY TRUE AND HOW MANY FALSE VALUES IN EACH COLUMN
df1.isna().sum()
```

```python
#  TO UDENTIFY WHETHER DATAFRAME HAS NULL VALUES OR NOT
df1.isna().any()
```

```python
# DROPPING ALL NULL VALUE

# df1.dropna()
df1
```

```python
# IF NON NULL VALUES IS LESS THAN 2 THEN DELTE THAT ROW
df1.dropna(thresh=3)
```

```python
# FILLING THE MISSING DATA
```

```python
# TO REPLACE/FILL NULL VALUES WITH 0
df1.fillna(0)
```

```python
# TO REPLACE/FILL EACH COLUMN HAVING NULL VALUES WITH PARTIULCAR VALUES
values={'A':0,'B':7,'C':8,'D':10}
df1.fillna(value=values)
```

```python
# .isna() for checking .dropna() for dropping null values .fillna for fillling null value
# fill column wise and drop is row wise
```

```python
df1.fillna(df1.mean())
```
