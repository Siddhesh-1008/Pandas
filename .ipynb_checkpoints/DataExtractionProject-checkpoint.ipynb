# FEATURE EXTRACTION

Make a new column for episode count

```python
# First EXTRACT THE EPISODE FROM TITLE AS EPISODE IS IN TITLE NOTATED WITH () EG:-(64PS)
df.loc[0]['Title']
```

```python
# EXTRACT EPISODES FROM TITLE COLUMN
def extract_epsiodes(text):
    start=text.find("(")
    end=text.find(")")
    return text[start+1:end]
df['Episodes']=df['Title'].apply(extract_epsiodes) 
df 
# # REPLACE EPS WITH EMPTY STRING 
df['Episodes']=df['Episodes'].str.replace(' eps',"")
df         
# # CONVERT STRING (64) INTO NUMERICAL
df['Episodes']=df['Episodes'].astype(int)
df
```

# Make a new column for time stamp

```python
df.head(5)
df.loc[2]['Title']
```

```python
def extract_ts(txt):
    start=txt.find(")");
    return txt[start+1:start+20]
df['Timestamp']=df["Title"].apply(extract_ts)
df['Timestamp']
df.head()
```

# Now wnat to count total number of months in that timestamp

```python
import pandas as pd

def count_months(txt):

    # AFTER DOING STRIP WE WILL HAVE start = "Oct 2022" AND end = "Dec 2022"
    start, end = txt.split("-")

    # pd.period_range(start, end, freq='M')
    # basically it calculates how many months are in between 
    # start.strip() basically remove unwanted space from start and end
    return len(pd.period_range(start.strip(),
                               end.strip(),
                               freq='M'))


df['Months_Count'] = df['Timestamp'].apply(count_months)
df
```
