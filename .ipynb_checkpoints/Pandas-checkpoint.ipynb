```python
# COMBINING MULTIPLE SERIES TO CREATE ONE DATAFRAME
```

```python
import pandas as pd 
import numpy as np
```

```python
data1 = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}

pd.DataFrame(data1)
```

```python
data2 = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
pd.DataFrame(data2)

# GIVE COLUMNS TITLE 
cols=["Name","Age","Country","Salary"]
data3=pd.DataFrame(data2,columns=cols)
data3
```

```python
# ACESSING PARTICULAR COLUMNS FROM DATAFRAME
```

```python
# SELECT SINGLE COLUMN AND THEN MULTIPLE COLUMNS
# IN DATAFRAMES INDEXING DOESNT WORKS ONLY COLUMN NAMES WORK
data3['Name']
```

```python
data3[["Name","Age"]]
```

```python
# ADDING NEW COLUMN IN 2D MATRIX
```

```python
data3["Designation"]=["Dr","Age","City","Salary"]
data3
```

```python
# DELETE PARTICULAR COLUMN FROM MATRIX BASED ON AXIS
```

```python
# axis=0 VERTICAL MEANS 0,1,2,3 
# axis=1 HORIZONTAL MEANS Name Age Country Salary Designation
data3.drop("Designation",axis=1)
```

```python
# TO DELETE COLUMNS PERMANENTY WE NEED TO WRITE inplace AS True
data3.drop("Designation",axis=1,inplace=True)
```

```python
data3
```

```python
# TO DELETE MULTIPLE COLUMNS PERMANENTLY JUST MAKE USE OF LIST
```

```python
# TO DELETE PARTICULAR ROW JUST MENTION INDEX
data3.drop(0,axis=0)
```
